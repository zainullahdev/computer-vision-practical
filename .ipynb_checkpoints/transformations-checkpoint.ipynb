```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../../opencv/J9KeYkEZf4HHD5LRGf799N.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()
```

```python
height,width = image.shape[:2]

move_x = 100.0
move_y = 0.0

T = np.float32([[1,0,move_x], [0,1,move_y]])
image_translated = cv2.warpAffine(image, T, (width,height))
image_translated[:,0:100] = (200,255,0)
plt.imshow(image_translated)
plt.tight_layout()
plt.show()
```

```python
# rotation

rotation_matrix = cv2.getRotationMatrix2D((width/2,height/2), 90, 1)

image_rotated = cv2.warpAffine(image, rotation_matrix, (width,height))
plt.imshow(image_rotated)
plt.tight_layout()
plt.show()
```

```python
image_rotated = cv2.transpose(image)
plt.imshow(np.fliplr(image_rotated))
plt.tight_layout()
plt.show()
```

```python
plt.imshow(image_rotated)
plt.tight_layout()
plt.show()
```

```python
image_resized = cv2.resize(image, None, fx=0.5, fy=0.5, interpolation= cv2.INTER_AREA)

plt.imshow(image_resized)
plt.tight_layout()
plt.show()
```

```python
image_smaller = cv2.pyrDown(image)

plt.imshow(image_smaller)
plt.tight_layout()
plt.show()
```

```python
image_larger = cv2.pyrUp(image)

plt.imshow(image_larger)
plt.tight_layout()
plt.show()
```

```python
# crop

image_cropped = image[100:350,220:500]

plt.imshow(image_cropped)
plt.tight_layout()
plt.show()
```

```python
# brightness

M = np.ones(image.shape, dtype='uint8') * 30
M[550:-1,440:500] += 150
image_brighter = cv2.add(image,M)
plt.imshow(image_brighter)
plt.show()
```

```python
# lion whitened 
image_cropped = image[100:350,220:500]
M = np.ones(image_cropped.shape, dtype='uint8') * 30
plt.imshow(cv2.add(image_cropped,M))
plt.tight_layout()
plt.show()
```

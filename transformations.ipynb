```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../../opencv/J9KeYkEZf4HHD5LRGf799N.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()
```

```python
height,width = image.shape[:2]

move_x = 100.0
move_y = 0.0

T = np.float32([[1,0,move_x], [0,1,move_y]])
image_translated = cv2.warpAffine(image, T, (width,height))
image_translated[:,0:100] = (200,255,0)
plt.imshow(image_translated)
plt.tight_layout()
plt.show()
```

```python
# rotation

rotation_matrix = cv2.getRotationMatrix2D((width/2,height/2), 90, 1)

image_rotated = cv2.warpAffine(image, rotation_matrix, (width,height))
plt.imshow(image_rotated)
plt.tight_layout()
plt.show()
```

```python
image_rotated = cv2.transpose(image)
plt.imshow(np.fliplr(image_rotated))
plt.tight_layout()
plt.show()
```

```python
img2 = cv2.imread('J9KeYkEZf4HHD5LRGf799N.jpg')

img2_copy = np.copy(img2)
b,g,r = cv2.split(img2_copy)

zeros = np.zeros(img2_copy.shape[:2], dtype='uint8')

plt.imshow(cv2.merge([r,zeros,zeros])) # plt imshow shows as rgb
plt.show()
```

```python

plt.imshow(cv2.merge([zeros,g,zeros])) # plt imshow shows as rgb
plt.show()
```

```python

plt.imshow(cv2.merge([zeros,zeros,b])) # plt imshow shows as rgb
plt.show()
```

```python

plt.imshow(cv2.merge([r,g,b])) # plt imshow shows as rgb
plt.show()
```

# Histogram

```python
colors = ['b','g','r']

for i in range(3):
    histogram = cv2.calcHist([img2_copy],[i],None,[256],[0,256])
    plt.plot(histogram,color = colors[i])
    plt.xlim([0,256])
plt.show()
```
